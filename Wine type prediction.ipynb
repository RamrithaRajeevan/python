```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l=['type','alcohol', 'malic_acid','ash','alcalinity_of_ash', 'magnesium', 'total_phenols','flavanoids','nonflavanoid_phenols', 'proanthocyanins','color_intensity','hue', 'od280/od315_of_diluted_wines', 'proline']
for i in l:
    df[i]=le.fit_transform(df[i])
df
```

```python
x=df.drop("type",axis=1)
x
```

```python
y=df['type']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
d=reg.predict(xtest)
d
```

```python
a1=accuracy_score(ytest,d)*100
a1
```

```python
d1=reg.predict(xtrain)
d1
```

```python
a2=accuracy_score(ytrain,d1)*100
a2
```

```python
v=[[65,49,18,1,21,68,91,9,31,56,50,102,92]]
w=reg.predict(v)
w
```

```python
if w==0:
    print("Class 1")
elif w==1:
    print("Class 2")
else:
    print("Class 3")
```
